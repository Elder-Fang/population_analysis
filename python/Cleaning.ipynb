The national growth factor is calculated as:

𝑅
national
=
Pop2030(UN-DESA)/Pop2017(UN-DESA)

We already have the total population for 2030 (PopTotal) and for 2017 from the tehsil data (sum of ALL_SEXES). Here's the code to calculate the national growth factor:


```python
# Display a sample of the 2030 population data for Pakistan
print("Sample of UN-DESA Population Data for 2030:")
print(age_group_data.loc[
    (age_group_data['Location'] == "Pakistan") &
    (age_group_data['Time'] == 2030)
])
```

```python
# Adjust the scale of UN-DESA PopTotal
age_group_data['PopTotal'] = age_group_data['PopTotal'] * 1000
print("Adjusted UN-DESA Population Data (2030):")
print(age_group_data.head())
```

```python
# Calculate total population in 2017 from Tehsil data
national_pop_2017 = tehsil_data['ALL_SEXES'].sum()

# Get total population in 2030 from Age Group data
national_pop_2030 = age_group_data.loc[
    (age_group_data['Location'] == "Pakistan") &
    (age_group_data['Time'] == 2030),
    'PopTotal'
].sum()

# Recalculate the National Growth Factor
national_growth_factor = national_pop_2030 / national_pop_2017

print(f"Adjusted National Population in 2030: {national_pop_2030}")
print(f"Adjusted National Growth Factor: {national_growth_factor}")
```

Adjusted National Population in 2030:
32,919,312 matches the total population for Pakistan in 2030 from the UN-DESA Age Group Data, specifically the "PopTotal" column for "Location: Pakistan" and "Time: 2030."

Adjusted National Growth Factor:
0.0217 is derived from dividing the total population in 2030 (32,919,312) by the total population in 2017 (sum of ALL_SEXES from the tehsil data). This value reflects the national growth rate between 2017 and 2030.

```python
# Calculate Admin-3 populations in 2030
tehsil_data['Pop_2030'] = tehsil_data['ALL_SEXES'] * national_growth_factor

# Display a sample of the updated Tehsil data with the 2030 population column
print("Sample of Admin-3 Data with 2030 Population Estimates:")
print(tehsil_data[['ADMN_UNIT', 'ALL_SEXES', 'Pop_2030']].head())

# Save the updated Tehsil data with 2030 population to a new CSV file for verification
tehsil_2030_file_path = "/content/drive/MyDrive/Unicef/Cleaned_data/all_tehsil_2030.csv"
tehsil_data.to_csv(tehsil_2030_file_path, index=False)
print(f"Tehsil data with 2030 population estimates saved to {tehsil_2030_file_path}")
```

Next Step:
We now calculate the number of children under 5 years old for each Admin-3 in 2030 using the proportion of children under 5 from the UN-DESA data. Here's the step-by-step plan:

Extract the proportion of children aged 0–4 in 2030 from the UN-DESA data.
Apply this proportion to the 2030 population (Pop_2030) for each Admin-3.

```python
# Extract the proportion of children aged 0–4 from UN-DESA data
# Get the proportion of children 0–4 in 2030 (PopTotal for 0–4 / Total Population)
age_group_population = age_group_data.loc[
    (age_group_data['Location'] == "Pakistan") &
    (age_group_data['Time'] == 2030) &
    (age_group_data['AgeGrp'] == "0-4"),
    'PopTotal'
].values[0]

total_population_2030 = age_group_data.loc[
    (age_group_data['Location'] == "Pakistan") &
    (age_group_data['Time'] == 2030),
    'PopTotal'
].sum()

proportion_under_5 = age_group_population / total_population_2030

print(f"Proportion of children under 5 in 2030: {proportion_under_5}")

# Calculate the number of children under 5 for each Admin-3 in 2030
tehsil_data['Children_Under_5_2030'] = tehsil_data['Pop_2030'] * proportion_under_5

# Display a sample of the updated Tehsil data
print("Sample of Admin-3 Data with Children Under 5 Estimates:")
print(tehsil_data[['ADMN_UNIT', 'Pop_2030', 'Children_Under_5_2030']].head())

# Save the updated Tehsil data with children under 5 estimates
children_under_5_file_path = "/content/drive/MyDrive/Unicef/Cleaned_data/all_tehsil_children_under_5_2030.csv"
tehsil_data.to_csv(children_under_5_file_path, index=False)
print(f"Tehsil data with children under 5 estimates saved to {children_under_5_file_path}")
#still need verifcation before it s counted correct
```
